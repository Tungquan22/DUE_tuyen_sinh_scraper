### Tách cột d1, d2, d3 thành cột môn thi & điểm

```python
def splitSubject(value):
    colon_pos = value.find(":")
    return value[:colon_pos]

def splitGrade(value):
    colon_pos = value.find(":")
    return value[(colon_pos + 1):]
```

```python
data["mon_1"] = data["d1"].apply(splitSubject)
data["diem_1"] = data["d1"].apply(splitGrade)
data["mon_2"] = data["d2"].apply(splitSubject)
data["diem_2"] = data["d2"].apply(splitGrade)
data["mon_3"] = data["d3"].apply(splitSubject)
data["diem_3"] = data["d3"].apply(splitGrade)
```

```python
data.diem_1.astype("float")
```

```python
data = data.astype({'diem_1': "float", 'diem_2': "float", 'diem_3': "float"})
```

```python
data.drop(['d1', 'd2', 'd3'], axis=1, inplace=True)
```
